```python
target_cols = ["상품 가격", "배송비", "결제 금액", "리뷰 점수"]
import matplotlib.pyplot as plt

for col in target_cols:
    col_data = merged_copy_renamed[col].dropna()

    # 사분위수 계산
    Q1 = col_data.quantile(0.25)
    Q3 = col_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 마스크 및 통계
    outlier_mask = (col_data < lower_bound) | (col_data > upper_bound)
    outlier_count = outlier_mask.sum()
    total_count = len(col_data)
    outlier_ratio = outlier_count / total_count

    # 결과 출력
    print(f"📊 [{col}]")
    print(f" - 전체 값 수: {total_count}")
    print(f" - 이상치 수: {outlier_count}")
    print(f" - 이상치 비율: {outlier_ratio:.2%}")
    if outlier_ratio < 0.05:
        print(" ✅ 이상치 비율 낮음 → 삭제해도 무방할 수 있음")
    else:
        print(" ⚠️  이상치 비율 높음 → 삭제 여부 검토 필요")

    # Boxplot 시각화
    plt.figure(figsize=(8, 1.5))
    plt.boxplot(col_data, vert=False)
    plt.title(f"Boxplot of '{col}'")
    plt.grid(True)
    plt.show()

    print("-" * 60)
```

# 만족도 점수 범주화 (3개 카테고리)  

```python
# 1~2점: 불만족(0), 3점: 보통(1), 4~5점: 만족(2)
def label_satisfaction(score):
    if score <= 2:
        return 0  # 불만족
    elif score == 3:
        return 1  # 보통
    else:
        return 2  # 만족

df_delivered['만족도_분류'] = df_delivered['리뷰 점수'].apply(label_satisfaction)
df_delivered['만족도_분류'].value_counts().sort_index()
```

    - 결제 금액 대비 배송비가 100% 이상인 경우는 대부분 특수 상황이므로 분석에서 제외함

    - 각 구간(빈)의 **실제 주문 건수(횟수)**를 나타냄

```python
# ✅ 3. 단순, 배송비를 절대적인 수치로 구분만 하는 것보다는,
# 결제 금액 대비 배송비가 차지하는 비중도 고려할 필요가 있다고 생각하여 비중과 관련된 새로운 칼럼을 추가.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 또는 사전 전처리된 데이터프레임 불러오기
# df = pd.read_csv("your_dataset.csv")

# 배송비 비중(%) 컬럼이 없다면 생성
if '배송비_비중_퍼센트' not in df_delivered_edited.columns:
    df_delivered_edited['배송비_비중_퍼센트'] = df_delivered_edited['배송비_비중'] * 100

# 100% 이하만 필터링 (이상치 제거) _ 특이 케이스
df_plot = df_delivered_edited[df_delivered_edited['배송비_비중_퍼센트'] <= 100]

# 여기서의 df_plot 은 '배송비_비중_퍼센트'] <= 100 인 칼럼들에 대한 dataframe

# 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df_plot['배송비_비중_퍼센트'], bins=50, kde=False)
plt.title("배송비 비중 분포 (X축: 0~100%)")
plt.xlabel("배송비 비중 (%)")
plt.ylabel("건수")
plt.xlim(0, 100)
plt.grid(True)
plt.tight_layout()
plt.show()
```

배송비 부담도

[실무 기준]
> | 배송비 비중 구간    | 배송비 등급 |
| ------------ | ------ |
| 0.00 \~ 0.05 | 매우 저렴  |
| 0.05 \~ 0.15 | 보통     |
| 0.15 \~ 0.30 | 다소 높음  |
| 0.30 초과      | 매우 부담  |


```python
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype


# 배송비 비중에 따라 배송비 등급 범주화 함수 정의
def classify_shipping_burden(ratio):
    if pd.isna(ratio):
        return np.nan
    elif ratio <= 0.03:
        return '부담 거의 없음'
    elif ratio <= 0.07:
        return '저부담'
    elif ratio <= 0.15:
        return '보통'
    elif ratio <= 0.30:
        return '약간 부담'
    else:
        return '고부담'

# 적용하여 새로운 컬럼 생성
df_delivered_edited['배송비_부담도'] = df_delivered_edited['배송비_비중'].apply(classify_shipping_burden)

# 확인
df_delivered_edited[['배송비_비중', '배송비_부담도']].head()



```
